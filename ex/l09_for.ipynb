```python
#Iterable

nums=[1,2,3,4,5]
it=iter(nums)
print(next(it))
print(next(it))
print(next(it))
print(next(it))
print(next(it)) 
# print(next(it)) 

it=iter(nums)
while True:
    try:
        n=next(it)
        print(n)
    except StopIteration:
        break
```

```python
#enumerate index+value
str="Hello"
print("인덱스가 필요한 경우")
for i in range(len(str)):
    print(i, str[i])    

print("enumerate 사용")
for i,ch in enumerate(str):
    print(i,ch)
```

```python
# 컴프리헨션
nums=[]
for i in range(3) :
    nums.append(i)
print(nums)


nums=[i for i in range(3) ]
print(nums)
```

```python
nums=[]
for i in range(3) :
    row=[]
    for j in range(2):
        row.append(i+j)
    nums.append(row)   

print(nums)

nums=[[i+j for j in range(2)] for i in range(3)]
print(nums)
```

```python
nums = [[None]*2 for _ in range(3)]
print(nums)

for i in range(3):
    for j in range(2):
        nums[i][j]=i+j
print(nums)
```

```python
# 컴프리헨션 with condition
nums=[ i for i in range(100) if i%2==1 ]
print(nums)
```

```python
# generator => 컴프리헨션 layzy (generator는 값을 한 번에 만들지 않고, 요청될 때 하나씩 만들어 주는 반복 가능한 객체로 iterable 하다)
gen=(i for i in range(100) if i%2==1 )
print(gen)
print(list(gen))
```
