```python
class User :
    # 생성자 self는 필드 접근자 this
    def __init__(self,name,age) :
        self.name = name
        self.age = age
    # 값을 비교하는 == 연산자
    def __eq__(self, other) :
        return self.name == other.name and self.age == other.age
    # toString
    def __repr__(self) :
        return f'User({self.name}, {self.age})'


u = User("경민코딩",39)
u2 = User("경민코딩",39)
print(u,u2)
print(u == u2)
```

```python
class Calc :
    def __init__(self, x, y) :
        self.x=x
        self.y=y
    def add(self) :
        return self.x+self.y

calc=Calc(10,20)
print( calc.add())
```

```python
#class 변수 (객체 생성 없이 사용 가능한 변수)
class Fruit :
    a = 'apple'
    b = 'banana'
    c = 'cherry'

print(Fruit.a,Fruit.b,Fruit.c)
```

```python
class Fruit :
    a = 'apple'
    b = 'banana'
    c = 'cherry'
    #class 변수로 무언가 하는 클래스 메소드
    @classmethod
    def show(cls) :
        print(cls.a,cls.b,cls.c)
    #객체 생성이 필요없는 클래스변수와 메소드와 동일하지만 클래스 변수 없이 메소드만 사용
    @staticmethod
    def show2() :
        print('static method')
    def __init__(self) :
        self.a="pineapple"
    def show3(self) :
        print(self.a)

Fruit.show()
Fruit.show2()
f=Fruit()
f.show3()
```

```python
#상속
class Animal :
    def __init__(self,name) :
        self.name=name
    def run(self) :
        print(f'{self.name} is running')

class Dog(Animal) :
    def __init__(self,name,breed) :
        super().__init__(name) #부모생성자 정의를  강제함
        self.breed=breed

    def run(self) :
        print(f'{self.name}({self.breed}) is running')

    def bark(self) :
        print(f'{self.name}({self.breed}) is barking')

ani=Animal("기린")
ani.run()

bichon=Dog("코에","비숑")
bichon.run()
bichon.bark()
```
