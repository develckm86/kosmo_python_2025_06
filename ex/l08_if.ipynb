```python
score=85
grade=None
if score>=90:
    grade='A'
elif score>=80:
    grade='B'
elif score>=70:
    grade='C'
elif score>=60:
    grade='D'
else:
    grade='F'
print(grade)
grade
```

```python
#--- 다중 조건 표현식(3항 연산자 이용) ---
score=75
grade = ('A' if score>=90 else
         'B' if score>=80 else
         'C' if score>=70 else
         'D' if score>=60 else
         'F')
# grade = 'A' if score>=90 else 'B' if score>=80 else 'C' if score>=70 else 'D' if score>=60 else 'F'

print(grade)
```

```python
# 3.10 이상에서만 지원되는 match 문법 활용
score = 75
match score:
    case s if s >= 90:
        grade = 'A'
    case s if s >= 80:
        grade = 'B'
    case s if s >= 70:
        grade = 'C'
    case s if s >= 60:
        grade = 'D'
    case _:
        grade = 'F'
print(grade)
```
