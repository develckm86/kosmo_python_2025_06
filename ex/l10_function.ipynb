# 함수 언패킹(Unpacking)은 파이썬 등에서 튜플, 리스트, 딕셔너리 같은 하나의 객체(컬렉션)에 묶여있는 값들을 꺼내어 여러 개의 개별 변수로 풀어주는 기능으로, 함수 호출 시 인자로 전달할 때 유용하며, *나 ** 연산자를 사용하여 인자 앞에 붙여서 사용합니다. 이는 여러 값을 묶는 패킹(Packing)과 반대되는 개념으로, 가변 인수(variable arguments)를 다룰 때 특히 유용합니다.


```python
# 가변 인자
def sum(*args) :
    print(args,type(args))
    total=0
    for n in args :
        total+=n
    return n

print(sum(10,20,30))
nums=[100,200,300]
# 개변인자로 전달
print(sum(*nums))
```

```python
# dict
def student_info(**student) :
    print(student,type(student))
    print(f"name={student['name']}")
    print(f"name={student['age']}")

student_info(name="경민",age=39)
stu={"name":"지형","age":28}
student_info(**stu)
```

```python
def student_say(name,age) :
    print(f"{name}의 나이는 {age} 입니다")

stu={"name":"지형","age":35}
student_say(**stu)
```

```python
a=100
def scope(n) :
    a=n
    return a

print(scope(10))
print(a)
```

```python
a = 100
def scope(n):
    global a
    a = n
    return a

print(scope(10))
print(a)
```
