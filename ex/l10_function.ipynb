```python
def sum(a,b,c) :
    print(a+b+c)
n=sum(10,20,30)
print(n)
```

```python
# list 언팩킹
def sum(*args) :
    print(args,type(args))
    total=0
    for n in args :
        total+=n
    return n

print(sum(10,20,30))
nums=[100,200,300]
# 개변인자로 전달
print(sum(*nums))
```

```python
def student_info(**student) :
    print(student,type(student))
    print(f"name={student['name']}")
    print(f"name={student['age']}")

student_info(name="경민",age=39)
```
