```python
a,b,c=10,20,-30
print("a + b =", a + b)
print("a - b =", a - b)
print("a * b =", a * b)
print("a / b =", a / b)
print("a // b =", a // b)
print("a % b =", a % b)
print("a ** b =", a ** b)

print("a == b =", a == b)
print("a != b =", a != b)
print("a > b =", a > b)
print("a < b =", a < b)
print("a >= b =", a >= b)
print("a <= b =", a <= b)

print("a is b =", a is b)
print("a is not b =", a is not b)
print("-a =", -a)
print("+a =", +a)
```

```python
print("abs(c) =", abs(c))
print("pow(a, 3) =", pow(a, 3))
print("divmod(a, 3) =", divmod(a, 3))

# 파이썬 round()는 반올림 규칙이 은행가 반올림(짝수 쪽으로) 이라는 특징이 있음.
print("round(2.5)=",round(2.5))  # 2
print("round(2.5)=",round(3.5))  # 4

print("bin(10)=",bin(10))  # '0b1010' 2진수
print("oct(10)=",oct(10))  # '0o12' 8진수
print("hex(10)=",hex(10))  # '0xa' 16진수
```

```python
import math
# 반올림 없음
print("sqrt(16) =", math.sqrt(16))          # 제곱근
print("ceil(3.2) =", math.ceil(3.2))        # 올림

print("floor(3.8) =", math.floor(3.8))      # 내림
print("trunc(3.9) =", math.trunc(3.9))      # 버림(소수점 제거)
```
