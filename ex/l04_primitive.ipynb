```python
str="abcde"
print(str, type(str))
print(str[1])  # indexing
# str[0]="z"  # 오류 발생, 문자열은 불변
```

```python
print(1==1)
print(1!=1)
b=True
print(b, type(b))
b=False
print(b, type(b))
```
