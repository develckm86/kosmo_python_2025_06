```python
# slicing
nums=[10,20,30,40,50,60,70,80,90]
print(nums[2:5])      # 2번 index부터 5번 index 전까지
print(nums[:4])       # 첫번째 index부터 4번 index 전까지
print(nums[5:])       # 5번 index부터 마지막 index까지
print(nums[-3:])      # 마지막에서 세번째 index부터 끝까지
print(nums[::2])      # 2칸씩 건너뛰기
print(nums[::-1])     # 역순으로
print(nums[1:9:3])    # 1번 index부터 7번 index 전까지 3칸씩 건너뛰기
print(nums[-1:-10:-2]) # 마지막 index부터 역순으로 2칸씩 건너뛰기
```

```python
copy_nums=nums[:]  # 리스트 전체 복사
print(copy_nums)
copy_nums[0]=999
print(copy_nums)
print(nums)  # 원본 리스트는 변경되지 않음
```

```python
a={1,2,3,4}
b={3,4,5,6}
print("집합 함수 사용")
print(a.union(b))          # 합집합
print(a.intersection(b))   # 교집합
print(a.difference(b))     # 차집합
print(b.difference(a))     # 차집합 
print("집합 연산자 사용")
print(a | b)          # 합집합
print(a & b)          # 교집합
print(a - b)          # 차집합
print(b - a)          # 차집합  
```
