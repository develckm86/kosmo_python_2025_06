```python
import requests

url="https://www.danawa.com"

header={"User-Agent":"Mozilla/5.0"} #파이썬이 아니라 브라우저로 접속하는 거야(거짓말)
#header : 문서의 정보(메타)를 포함하는곳 

res=requests.get(url=url,headers=header) 
# url 로 리소스를 요청하는 행위를 get 방식이라 부른다. (post : 양식제출 로그인, 회원가입, 설문지..)
print(res.text)
```

```python
import requests
from bs4 import BeautifulSoup as bs
# https://search.danawa.com/dsearch.php?query=%EC%9B%90%ED%94%BC%EC%8A%A4+%ED%94%BC%EA%B7%9C%EC%96%B4&tab=main
url = "https://search.danawa.com/dsearch.php"
params = {
    "query": "원피스 피규어",
    "tab":"main"
}

headers = {
    "User-Agent": "Mozilla/5.0"
}

res = requests.get(url, params=params, headers=headers)
print(res.url)    # 인코딩된 실제 요청 URL
soup=bs(res.text)
prod_main_infos=soup.select(".prod_main_info")
# print(prod_main_infos)
# print(len(prod_main_infos))
prods=[]
for prod in prod_main_infos:
    name_tag = prod.select_one(".prod_name a")
    img_tag = prod.select_one(".thumb_image img")
    price_tag = prod.select_one(".price_sect strong")
    cate_tag = prod.select_one(".prod_category_location dd a")

    img_url = img_tag["src"]
    if img_url.startswith("//"):
        img_url = "https:" + img_url

    item = {
        "name": name_tag.get_text(strip=True),
        "link": name_tag["href"],
        "image": img_url,
        "price": price_tag.get_text(strip=True) + "원" if price_tag else None,
        "category": cate_tag.get_text(" > ", strip=True) if cate_tag else None
    }

    prods.append(item)

print(prods)
```

```python
import pandas

pandas.DataFrame(prods)
```
