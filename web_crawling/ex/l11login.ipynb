# 원시적인 방법으로 크롤링 로그인!
- 원시적 : 크롤링하는 브라우저는 자동로그인이 불가능하다. 때문에 매번 로그인을 해줘야한다.
- 로그인할 때 주의할 점 : 크롤링 로그인을 막는 사이트가 유난히 많다.

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

#막히거나 안되는 거 부터 경험!!

url="https://nid.naver.com/" 
driver=webdriver.Chrome()
driver.get(url)
userId="8ozmagazine"
userPw="chlrudals86"
idInput=driver.find_element(By.ID,"id")
pwInput=driver.find_element(By.ID,"pw")
loginBtn=driver.find_element(By.ID,"log.login")

#idInput.send_keys(userId)
#pwInput.send_keys(userPw)
#loginBtn.click()

#네이버 서버가 컴퓨터인것을 속이려면...
#send_keys가 입력하는 속도가 너무 빠르다. (클립보드에 복사해서 붙여넣기)
#모든 것을 입력하고 로그인 버튼을 누르는 것 까지가 너무 빠르다. (tiem.sleep)
```

```python
import pyperclip #클립보드를 사용하는 라이브러리
import time
time.sleep(1)
pyperclip.copy(userId)
idInput.send_keys(Keys.CONTROL+"v")
time.sleep(2)
pyperclip.copy(userPw)
pwInput.send_keys(Keys.CONTROL+"v")
time.sleep(2)
loginBtn.click()
#붙여넣기 control+v(window) or command+v(맥)


```
