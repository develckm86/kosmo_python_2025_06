# beatifulsoup4 라이브러리 
- http 통신으로 받아온 text 형식의 html 문서를 dom 형식으로 파싱한다.
- text를 html로 바꾼다.
- dom tree 검색이 가능 node 검색이 가능!!
- pip install beautifulsoup4 
- (맥)만약 오류 ->  sudo  python3 -m pip install bs4

```python
import requests as req # as : 모듈 전체에 가짜이름 
from bs4 import BeautifulSoup as bs # from import : 특정 함수 또는 모듈을 가져오는 방법

url="https://www.danawa.com"
header={"User-Agent":"Mozilla/5.0"}

res=req.get(url=url,headers=header)

res.text
```

```python
soup=bs(res.text,"html.parser") # text 를 html dom 으로 파싱하겠다.
#파싱 : 변환되지 않을 수도 있는 것을 변환하는 행위                  "13ㅁ"-> 13 x 
# print(soup.prettify())
```

```python
#navs=soup.select("#danawa_menubar>.main-header__nav>.link_nav") # 복수의 요소(노드)를 찾는 것 (리스트 반환)
# 구체적인 선택자를 사용하는 이유!
# 1. 중복이 존재할 수 있어서
# 2. 가독성 
# 3. 선택자가 구체적일 수록 동작이 더 빠르다. 
navs=soup.select(".link_nav") # 복수의 요소(노드)를 찾는 것 (리스트 반환)
# nav 이 텍스트 노드와 href 를 출력하세요! 

for nav in navs :
    print(f"{nav.text} / {nav.get('href')}")
```

```python
#움직이는 메뉴의 선택자를 찾아서 출력하세요
moveNavs=soup.select(".main-top100__nav>.main-top100__nav__anchor")
for nav in moveNavs :
    print(nav.text)
    print(nav.get("href")) #get(속성) : 해당 노드의 속성 값을 반환
    #href 가 아이디를 참조하면 해당 아이디로 스크롤 이동한다.
```
