```python
import time
time.sleep(2)

while True :
    prevHeight=driver.execute_script("return document.body.scrollHeight") #스크롤 내리기 전

    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    time.sleep(2)

    nowHeight=driver.execute_script("return document.body.scrollHeight") #스크롤 내린 후
    
    if nowHeight == prevHeight :
        break

```

```python
from bs4 import BeautifulSoup

body=driver.find_element(By.TAG_NAME,"body")
body.send_keys(Keys.HOME)
time.sleep(1) # 스크롤이 이동후 요소 숫자가 변경되는데 시간이 조금 걸림
imgs=[]
soup=BeautifulSoup(driver.page_source,"html.parser") #lxml lib
imgs+=soup.select("._aagv>img")
driver.execute_script("document.querySelector('._ac7v:last-child').setAttribute('id','lastBlock')")
print(len(imgs))

while True :

    prevScrollY=driver.execute_script("return window.scrollY")
    body.send_keys(Keys.DOWN) #스크롤을 조금씩 내림
    time.sleep(0.5)
    nowScrollY=driver.execute_script("return window.scrollY") 
    #스크롤을 내렸는데도 위치변화가 없으면 끝에 있다는 뜻 (반복문 끝)

    try : 
        driver.find_element(By.ID,"lastBlock") #없으면 찾을 수 없음 오류가 발생
        #오류가 발생했다는 것은 그 시점부터 처음 보는 이미지 리스트들(수집)
    except :
        soup=BeautifulSoup(driver.page_source,"html.parser") #lxml lib
        imgs+=soup.select("._aagv>img")
        driver.execute_script("document.querySelector('._ac7v:last-child').setAttribute('id','lastBlock')")

    if prevScrollY == nowScrollY :
        soup=BeautifulSoup(driver.page_source,"html.parser") #lxml lib
        imgs+=soup.select("#lastBlock~._ac7v>div>a>._aagu>._aagv>img")
        break

print(len(imgs))

```

```python
import os
from urllib.request import urlretrieve

imgDir="C:\\Users\\USER\\web_crawling_study\\imgs\\eunwo.o_c"

if not os.path.isdir(imgDir) :
    os.mkdir(imgDir)

i=0

for img in imgs :
    i+=1
    src=img.get("src")
    time.sleep(0.005)
    #너무 빠르게 이미지를 다운 받으면 크롤링으로 인식해서 오류(10060) 발생
    urlretrieve(src,imgDir+"\\차은우"+str(i)+".jpg")



```
