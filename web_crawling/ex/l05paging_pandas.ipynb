```python
import requests
import pyautogui #gui 키보드와 마우스를 자동으로 움직이는 라이브러리 + 창 (alert,prompt,password,confirm)
from bs4 import BeautifulSoup

query=pyautogui.prompt("검색어를 입력") #맥에서 종료가 안되는 오류가 존재함.. 

titles=[]
pointNums=[]
pointCnts=[]
princes=[]

for page in range(1,11) :
    url=f"https://search.danawa.com/dsearch.php?query={query}&page={page}&limit=40"
    headers={"User-Agent":"Mozilla/5.0"}
    response=requests.get(url=url,headers=headers)
    soup=BeautifulSoup(response.text,"html.parser")
    #http 통신이 옛날은 html만 주고받음, 하지만 지금은 모든 파일을 받는다.
    print(f"#######################{page}페이지###############################")
    prods=soup.select(".prod_main_info") #검색한 아이템 정보
    

    for prod in prods :
        title=prod.select_one(".click_log_product_standard_title_")
        if title != None :
            titles.append(title.text)
        else :
            titles.append("제목없음")

        pointNum=prod.select_one(".point_num>strong")
        pointCnt=prod.select_one(".click_log_prod_content_count>strong")
        #리뷰점수가 존재해야 리뷰수가 존재
        if pointNum!=None :
            pointNums.append(pointNum.text)
            pointCnts.append(pointCnt.text)
        else :
            pointNums.append(0)
            pointCnts.append(0)

        price=prod.select_one(".click_log_product_standard_price_>strong")

        if price !=None :
            princes.append(price.text)
        else :
            princes.append(0)
        
```

```python
import pandas
data={"상품명":titles,"가격":princes,"별점":pointNums,"리뷰":pointCnts} #딕셔너리
result=pandas.DataFrame(data=data)
result
```
