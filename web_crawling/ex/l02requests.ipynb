# 요청과 응답
- 브라우저에서 url로 리소스를 요청하는 것을 request라하고 요청한 리소스를 응답하는 것을 response라 부른다.
- 브라우저는 기본적으로 비연결성 통신(이메일,문자,http)을 한다.(연결성 통신 : 소켓,전화,채팅) 예외적으로 웹소켓을 지원하기도 한다.
- html 문서를 요청하는 행위가 request 
- 요청한 문서를 받는 것을 response (상태status code가 존재 200 성공, 300 서버 내의 다른 리소스로 이동,400 존재하지 않음(클라이언트의 잘못), 500 서버내 오류)

# pip install requests 
- python 으로 http 리소스를 요청하는 라이브러리
- 웹크롤링 : 자동으로 웹상의 데이터를 수집 또는 접속 (데이터 == 돈 : 크롤링 데이터 도둑)
- 데이터 스크래핑 : 전문적으로 웹상의 데이터를 수집 저장
- 공공 API OR API : 국가 OR 회사 OR 개인이 수집한 데이터나 어프리케이션을 사용하는 권한을 제공
- 예: 항공기 검생 어플 
- 1. 웹크롤링해서 스카이스캐너 같은 회사에서 항공기 내역을 수집 
- 2. 공공 API에서 예약가능한 항공기내역 수집

```python
import requests as req

url="https://www.megabox.co.kr/"
header={"User-Agent":"Mozilla/5.0"} #파이썬이 아니라 브라우저로 접속하는 거야(거짓말)
#header : 문서의 정보(메타)를 포함하는곳 
# url="https://www.naver.com" 
res=req.get(url=url,headers=header) 
# url 로 리소스를 요청하는 행위를 get 방식이라 부른다. (post : 양식제출 로그인, 회원가입, 설문지..)
print(res.text)
```
