- 예전의 자바스크립트는 요소를 움직이기거나서 사라지게 하는 것에 그쳤다면 현재는 통신을 주로 하고 있기 때문에(AJAX비동기식통신 : 화면전환 없이 통신) requests 라이브러리로 웹크롤링을 완벽하게 하는 것은 무리가 있다.(처음에 로딩한 페이지만 크롤링 가능)

```python
from selenium import webdriver as wd 
from selenium.webdriver.common.by import By #dom 선택자를 어떻게 할건지를 고를때 사용(상수)
from selenium.webdriver.common.keys import Keys #키보드의 특정 키를 누를때 사용
import time 

driver=wd.Chrome() #Chrome 브라우저를 제어하기 위해 드라이버를 가져옴
driver.get("https://www.naver.com") #html dom 을 반환 
time.sleep(2)
query=driver.find_element(By.ID,"query") #bs4== select_one(#query)
query.send_keys("lg 세탁기") #너무 빠르게 글씨를 써도 컴퓨터로 인식함 (클립보드에 복사 붙여넣기)
time.sleep(1)
query.send_keys(Keys.ENTER)
```
