# 유튜브나 인스타그램에서 무한 스크롤로 불러오는 데이터를 크롤링해 보자
- 무한 스크롤 == 스크롤 페이징 (비동기식 통신)
- 스크롤이 내려가는 것을 자바스크립트가 인식해서 컨텐츠를 페이징 순서대로 불러온다.

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup


driver=webdriver.Chrome()
url="https://www.youtube.com/"
driver.get(url)
searchInput=driver.find_element(By.CSS_SELECTOR,"input#search")
searchBtn=driver.find_element(By.CSS_SELECTOR,"button#search-icon-legacy")
searchInput.send_keys("이연복의 복주머니")
searchBtn.click()
#################### 검색 후 다른 페이지로 이동 
driver.implicitly_wait(10) #다른 페이지 로딩까지 최대 10초까지 기다림
mainLink=driver.find_element(By.ID,"main-link")
mainLink.click()
#####################
driver.implicitly_wait(10)
tab=driver.find_element(By.CSS_SELECTOR,"[tab-title=\"동영상\"]")
tab.click()
###################### 동영상 무한 스크롤 로딩
driver.implicitly_wait(10)
body=driver.find_element(By.TAG_NAME,"body") #body는 오직 한개 뿐
for i in range(5) :
    body.send_keys(Keys.END)
    time.sleep(1)

soup=BeautifulSoup(driver.page_source,"html.parser")

views=soup.select("#metadata-line>.inline-metadata-item.style-scope.ytd-video-meta-block")

#views=soup.select("#metadata-line>.inline-metadata-item:first-child") #조회수
#views=soup.select("#metadata-line>.inline-metadata-item:last-child") #등록일
#select_one  첫번째 조회수 빼기 
#긴 문자열에서 회를 기준으로 앞 뒤로 잘라서 사용
for v in views :
    print(v.text)
```
