```python
import requests as req
from bs4 import BeautifulSoup as bs

query=input("무엇을 검색하시겠습니까?")
url=f"https://search.danawa.com/dsearch.php?query={query}"
headers={"User-Agent":"Mozilla/5.0"}
res=req.get(url=url,headers=headers)
soup=bs(res.text,"html.parser")
# print(soup.prettify())
```

```python
# prices=soup.select(".rank_one>.price_sect>.click_log_product_standard_price_")
# for p in prices :
#     print(p.text)
items=soup.select(".prod_main_info")
for item in items :
    title=item.select_one(".click_log_product_standard_title_").text
     #select_one() 여러개 찾아도 처음 찾는 1개만 반환
    price=item.select_one(".rank_one>.price_sect>.click_log_product_standard_price_").text
    point=item.select_one(".point_num")
    #None.text 오류 
    if(point==None):
        point="리뷰없음"
    else :
        point=point.text

    price=price.replace(",","")
    price=price.replace("원","")
    print(f"{title}\{price}원\{point}")
    #price -> (, 원을 제외하고) 정수로 변환하세요~ 
```
