## UTF-8에서 “문자 하나의 시작”을 알 수 있는 방법

UTF-8은 가변 길이의 인코딩으로 문자 하나가 1~4바이트까지 차지한다.

UTF-8의 규칙

각 바이트 앞부분 비트 패턴을 보면 “이 바이트가 새 문자 시작인지” 알 수 있음.

바이트 타입	비트 패턴	의미
ASCII	0xxxxxxx	문자 1바이트짜리, 시작
첫 바이트(2바이트 문자)	110xxxxx	새 문자 시작
첫 바이트(3바이트 문자)	1110xxxx	새 문자 시작
첫 바이트(4바이트 문자)	11110xxx	새 문자 시작
이어지는 바이트	10xxxxxx	문자 내부 바이트 (시작 아님)

236 → 0xEC → 11101100
235 → 0xEB → 11101011
